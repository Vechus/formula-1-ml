```python
import pandas as pd
import numpy as np
```

```python
races = pd.read_csv('data/races.csv')
results = pd.read_csv('data/results.csv')
driver_standings = pd.read_csv('data/driver_standings.csv')
constructor_standings = pd.read_csv('data/constructor_standings.csv')
weather = pd.read_csv('data/weather.csv')

print('races: ', races.shape)
print('results: ', results.shape)
print('ds: ', driver_standings.shape)
print('cs: ', constructor_standings.shape)
print('weather: ', weather.shape)
```

```python
races.head()
```

```python
results.drop('Time', inplace=True, axis=1)
results.head()
```

```python
driver_standings.drop(['driver_points_after_race', 'driver_wins_after_race', 'driver_standings_pos_after_race'] ,axis = 1, inplace = True)
driver_standings.rename(columns={'driver': 'driverID'}, inplace=True)
driver_standings.head()
```

```python
constructor_standings.drop(['constructor_points_after_race', 'constructor_wins_after_race','constructor_standings_pos_after_race' ],axis = 1, inplace = True)
constructor_standings.rename(columns={'constructor': 'constructorID'}, inplace=True)
constructor_standings.head()
```

```python
weather.head()
```

```python
df1 = pd.merge(races, weather, how='inner', on=['season', 'round', 'circuitID']).drop(['url', 'date', 'raceName', 'weather'], axis=1)
df1.head()
```

```python
df2 = pd.merge(df1, results, how='inner', on=['season', 'round'])
df2.head()
```

```python
df3 = pd.merge(df2, driver_standings, how='left', on=['season', 'round', 'driverID'])
df3.head()
```

```python
df4 = pd.merge(df3, constructor_standings, how='left', on=['season', 'round', 'constructorID'])
df4.head()
```

```python
# fill null values with 0 only for feature that are probably miscalculated, drop other nans
for col in ['driver_points', 'driver_wins', 'driver_standings_pos', 'constructor_points', 'constructor_wins', 'constructor_standings_pos']:
    df4[col] = df4[col].fillna(0)
    df4[col] = df4[col].astype(int)

df4.dropna(inplace=True)
```

```python
df4.drop(['driver', 'constructor', 'number'], axis=1, inplace=True)
```

```python
# convert to boolean

for col in ['weather_warm', 'weather_cold','weather_dry', 'weather_wet', 'weather_cloudy']:
    df4[col] = df4[col].map(lambda x: bool(x))
```

```python
df4.head()
```

```python
df4.to_csv('data/processed_data.csv', index=False)
```
